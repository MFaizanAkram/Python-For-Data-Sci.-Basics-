# Pakistan Data

```python
# Dropping unnecessary columns.
data.columns
```

```python
new_data=data.drop(['Domain Code', 'Domain', 'Area Code (FAO)', 'Element Code','Element', 'Item Code (FAO)', 'Year Code', 'Unit','Flag', 'Flag Description'], axis=1)
new_data
```

```python
pak=new_data[new_data['Area']=='Pakistan'].groupby(["Year", "Item", "Value"]).mean()
pak
```

```python
# Saving csv file
pak.to_csv('pak.csv')
```
